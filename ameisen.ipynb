```python
from ipywidgets import interact
import matplotlib.pyplot as plt
```

```python
def berechne_neue_position(x, y, richtung):
    if richtung == 'w':
        y = y + 1
    if richtung == 's':
        y = y - 1
    if richtung == 'd':
        x = x + 1
    if richtung == 'a':
        x = x - 1
    if richtung == 'e':
        x = x + 1
        y = y + 1
    if richtung == 'q':
        x = x - 1
        y = y + 1
    if richtung == 'y':
        x = x - 1
        y = y - 1
    if richtung == 'x':
        x = x + 1
        y = y - 1
    return x, y
```

```python
def f(richtungen):

    x = 0
    y = 0
    
    alle_x = [x]
    alle_y = [y]

    for richtung in richtungen:
        x, y = berechne_neue_position(x, y, richtung)
        alle_x.append(x)
        alle_y.append(y)
    
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.axis('square')
    
    plt.plot(alle_x, alle_y)
    plt.plot(x, y, 'o')
```

```python
test = 'aawwwwwddsasdsasds'
f(test)
```

```python
haus_vom_nikolaus = 'aaeeaaxxwwqyss'
f(haus_vom_nikolaus)
```

```python
interact(f, richtungen='');
```

```python
from random import choices

richtungen = choices('wasdqeyx', k=50)
f(richtungen)
```
